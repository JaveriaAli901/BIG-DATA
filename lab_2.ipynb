**Python Data Structures**
**practice lists, dictionaries, tuples, and sets**

```python
# List
grades = [85, 90, 78, 92]
print("List of grades:", grades)

# Tuple
student_info = ("Ali", 18, "A")
print("Tuple:", student_info)

# Dictionary
student = {"name": "Ali", "age": 18, "grade": "A"}
print("Dictionary:", student)

# Set
unique_grades = {85, 90, 78, 92, 85}
print("Set (unique values only):", unique_grades)
```

**Functions for Data** **Processing**

```python
def calculate_average(grades_list):
    return sum(grades_list) / len(grades_list)

grades = [85, 90, 78, 92]
avg = calculate_average(grades)
print("Average grade:", avg)
```

**CSV Files**

```python
import pandas as pd

# Step 1: Read the Excel file
df = pd.read_excel('/content/student_grades.xlsx')

# Step 2: Check the data (optional)
print("Original Excel data:")
print(df)

# Step 3: Save it as a CSV file
df.to_csv('/content/student_grades.csv', index=False)

print("Excel file has been converted to CSV successfully!")
```

```python
import csv

# Reading CSV
with open('student_grades.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)

# Writing CSV
with open('new_grades.csv', 'w', newline='') as file:
    fieldnames = ['name', 'grade']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerow({'name': 'Ali', 'grade': 90})
```

**JSON Files**

```python
import json

# Writing JSON
data = {"students": [{"name": "Ali", "grade": 90}, {"name": "Sara", "grade": 85}]}
with open('students.json', 'w') as file:
    json.dump(data, file)

# Reading JSON
with open('students.json', 'r') as file:
    loaded_data = json.load(file)
    print(loaded_data)
```

**Text Files**

```python

average_grade = 85.5
top_students = ['Ali', 'Sara']


with open('results.txt', 'w') as file:
    file.write(f"Average grade of students: {average_grade}\n")
    file.write(f"Top students (grade >= 90): {', '.join(top_students)}\n")

with open('results.txt', 'r') as file:
    content = file.read()
    print(content)
```

**List**

```python
grades = [85, 90, 78, 92]

new_grades = [g + 5 for g in grades]
print(new_grades)


students = [{"name": "Ali", "grade": 95}, {"name": "Sara", "grade": 85}]
top_students = [s['name'] for s in students if s['grade'] >= 90]
print(top_students)
```

**Handle Errors and Exceptions**

```python
grades = [85, 90, 'A', 92]

def safe_average(lst):
    try:
        return sum(lst) / len(lst)
    except TypeError:
        print("Error: List contains non-numeric values.")

print(safe_average(grades))
```
