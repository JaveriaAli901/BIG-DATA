```python
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
df = pd.read_csv(url)

df.head()
```

```python
df = df.dropna()

df.columns = ["Country", "Code", "Year", "Population"]

df.head()
```

```python
pakistan_data = df[df["Country"] == "Pakistan"]

line_fig = px.line(pakistan_data,
                   x="Year",
                   y="Population",
                   title="Population Growth of Pakistan")

line_fig.show()
```

```python
latest_year = df["Year"].max()

top10 = df[df["Year"] == latest_year] \
        .sort_values(by="Population", ascending=False) \
        .head(10)

bar_fig = px.bar(top10,
                 x="Country",
                 y="Population",
                 title="Top 10 Countries by Population")

bar_fig.show()
```

```python
scatter_data = df[df["Year"] == 2020]

scatter_fig = px.scatter(scatter_data,
                         x="Population",
                         y="Year",
                         size="Population",
                         hover_name="Country",
                         title="Scatter Plot with Hover")

scatter_fig.show()
```

```python
map_data = df[df["Year"] == 2020]

map_fig = px.choropleth(map_data,
                        locations="Code",
                        color="Population",
                        hover_name="Country",
                        title="World Population Map")

map_fig.show()
```

```python
sample = df[df["Year"] == 2020].head(50)

fig3d = px.scatter_3d(sample,
                      x="Population",
                      y="Year",
                      z="Population",
                      color="Country",
                      title="3D Scatter Plot")

fig3d.show()
```

```python
anim_fig = px.scatter(df,
                      x="Population",
                      y="Country",
                      size="Population",
                      animation_frame="Year",
                      title="Population Over Time Animation")

anim_fig.show()
```

```python
dashboard = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Line Chart", "Bar Chart", "Scatter Plot", "Map"),
    specs=[[{"type": "xy"}, {"type": "xy"}],
           [{"type": "xy"}, {"type": "choropleth"}]]
)

# Line chart
for trace in line_fig.data:
    dashboard.add_trace(trace, row=1, col=1)

# Bar chart
for trace in bar_fig.data:
    dashboard.add_trace(trace, row=1, col=2)

# Scatter plot
for trace in scatter_fig.data:
    dashboard.add_trace(trace, row=2, col=1)

# Map (special handling)
for trace in map_fig.data:
    dashboard.add_trace(trace, row=2, col=2)

dashboard.update_layout(
    height=800,
    title_text="Global Population Dashboard"
)

dashboard.show()
```

```python
dashboard.write_html("dashboard.html")

print("✅ File saved as dashboard.html")
```
