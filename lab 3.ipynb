```python

import numpy as n
import pandas as pd
import time

print("NumPy Version:", np.__version__)
```

```python
# Generate synthetic stock price data (100 days)
np.random.seed(42)

days = np.arange(1, 101)
prices = np.random.normal(loc=100, scale=10, size=100)

# Create DataFrame
df = pd.DataFrame({
    "Day": days,
    "Stock Price": prices
})

print(df.head())
```

```python
# 1D Array
arr_1d = np.array([1, 2, 3, 4, 5])

# 2D Array
arr_2d = np.array([[1, 2, 3],
                   [4, 5, 6]])

# 3D Array
arr_3d = np.array([[[1, 2], [3, 4]],
                   [[5, 6], [7, 8]]])

print("1D:\n", arr_1d)
print("2D:\n", arr_2d)
print("3D:\n", arr_3d)
```

```python
# Indexing
print("First element:", arr_1d[0])
print("Element from 2D:", arr_2d[1, 2])

# Slicing
print("Slice:", arr_1d[1:4])

# Reshaping
reshaped = np.arange(12).reshape(3, 4)
print("Reshaped Array:\n", reshaped)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Element-wise
print("Addition:", a + b)
print("Multiplication:", a * b)

# Broadcasting
print("Broadcast Add:", a + 10)
```

```python
data = prices

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("25th Percentile:", np.percentile(data, 25))
print("75th Percentile:", np.percentile(data, 75))
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Dot Product
print("Dot Product:\n", np.dot(A, B))

# Matrix Multiplication
print("Matrix Multiplication:\n", A @ B)
```

```python
# Random numbers
rand_nums = np.random.rand(5)
print("Random Numbers:", rand_nums)

# Random integers
rand_ints = np.random.randint(1, 100, size=5)
print("Random Integers:", rand_ints)

# Sampling from dataset
sample = np.random.choice(prices, size=10)
print("Sample:", sample)
```
